```python
#Import data
df_uber = pd.read_csv("UberData.csv")
df_uber.head()
```

```python
# set marginal cost
c = 10
```

```python
# 1C: Estimating the demand function

demand_model = OLS.from_formula('AcceptedRides ~ 1 + Price + AppOpens', df_uber)
fitted_model = demand_model.fit()
print(fitted_model.summary)

# store the three alpha coefficients for use later.
alpha_0 = fitted_model.params.Intercept
alpha_1 = fitted_model.params.Price # finish this
alpha_2 = fitted_model.params.AppOpens # finish this
```

```python
# 1D: Prices that maximize social surplus

is_capacity_binding = 1. * (df_uber['Price'] > c) # What is the condition for the capacity constraint to bind? Note the 1. * () converts from Boolean (True/False) to a float (a number)

equilibrium_price_if_binding = is_capacity_binding*(df_uber['AcceptedRides'] - alpha_0 - alpha_2*df_uber['AppOpens'])/alpha_1 # finish this

equilibrium_price = (is_capacity_binding==0)*c  + (is_capacity_binding==1)*equilibrium_price_if_binding # finish this

equilibrium_price.describe()
```
